# Neurons in keras

```python
import keras as kr
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']= (10, 10)
```

## Linear

```python
# create a new neural network
m = kr.models.Sequential()

# add a single neuron into a single layer, initialized with weight 1 and bias 0
m.add(kr.layers.Dense(1, input_dim=1, activation="linear", kernel_initializer="ones", bias_initializer="zeros"))

# compile the model
m.compile(loss="mean_squared_error", optimizer="sgd")
```

```python
# create input values
x = np.arange(0.0, 10.0, 1)

# run each x value through neural network
y = m.predict(x)
```

```python
plt.plot(x, y, 'k.')
```

## Single Linear Neuron

```python
# create new neural network
m = kr.models.Sequential()

# add a single neuron in a single layer, initialized with weight 2 and bias 1
m.add(kr.layers.Dense(1, input_dim=1, activation="linear", kernel_initializer=kr.initializers.Constant(value=2), bias_initializer=kr.initializers.Constant(value=1)))

# compile the model
m.compile(loss="mean_squared_error", optimizer="sgd")
```

```python
# create some input values
x = np.arange(0.0, 10.0, 1)

# run each x value through the neural network
y = m.predict(x)
```

```python
# plot the values
plt.plot(x, y, 'k.')
```

## Two Nuerons

```python
# create a new neural network
m = kr.models.Sequential()

# add 2 neurons in a single layer
m.add(kr.layers.Dense(2, input_dim=1, activation="linear"))

# add a single neuron in a single layer, initialized with weight 1 and bias 0
m.add(kr.layers.Dense(1, activation="linear", kernel_initializer=kr.initializers.Constant(value=1), bias_initializer=kr.initializers.Constant(0)))

# Set the weight/bias of the 2 neurons
m.layers[0].set_weights([np.matrix([2,3]), np.array([-5,-3])])

# compile the model
m.compile(loss="mean_squared_error", optimizer="sgd")
```

```python
# create some inputs
x = np.arange(0.0, 10.0, 1)

# run each x value through neural network
y = m.predict(x)
```

```python
# plot values
plt.plot(x, y, 'k.')
```

## Sigmoid

```python
# create a new neural network
m = kr.models.Sequential()

# add a single neuron in a single layer, initialized with weight 1 and bias 0, with sigmoid activation
m.add(kr.layers.Dense(1, input_dim=1, activation="sigmoid", kernel_initializer=kr.initializers.Constant(value=1), bias_initializer=kr.initializers.Constant(value=0)))

# compile model
m.compile(loss="mean_squared_error", optimizer="sgd")
```

```python
# create some inputs
x = np.arange(-10.0, 7.5, 1)

# run each x value through neural network
y = m.predict(x)
```

```python
# plot values
plt.plot(x, y, 'k.')
```
